```r
library(httr)
library(jsonlite)
```

```r
# using jsonlite directly
url_movie_1 <- "http://www.omdbapi.com/?apikey=a79b2c95&i=tt0076759"
movie1.details <- fromJSON(url_movie_1)
movie1.details
```

```r
# using httr & jsonlite

movie_2 <- httr::GET(url = "http://www.omdbapi.com/?apikey=a79b2c95&i=tt0076759")
movie_2_json <- httr::content(movie_2, as = "text")
movie_2_json_httr <- fromJSON(movie_2_json)
movie_2_json_httr
```

```r
# testing parameters for title search using multiple words

base_search_url <- "http://www.omdbapi.com/?s="
apikey <- "apikey=a79b2c95"
search <- "guardians+galaxy"

#http://www.omdbapi.com/?s=guardians+3+galaxy&apikey=a79b2c95

search_url_1 <- paste0(base_search_url, paste(search, apikey, sep = '&'))

search_1 <- httr::GET(url = search_url_1)
search_1_httr <- httr::content(search_1, as = "text")
search_1_httr_json <- fromJSON(search_1_httr)
search_1_httr_json
```

```r
# get imdb id of the first result
search_1_httr_json$Search$imdbID[1]
```

```r
imdb_id <- search_1_httr_json$Search$imdbID[1]
imdb_id
```

```r
# get movie details from IMDB ID

id_search_url <- "http://www.omdbapi.com/?i="
apikey <- "apikey=a79b2c95"
id <- imdb_id

#http://www.omdbapi.com/?i=tt2015381&apikey=a79b2c95

id_url_1 <- paste0(id_search_url, paste(id, apikey, sep = '&'))

id_1 <- httr::GET(url = id_url_1)
id_1_httr <- httr::content(id_1, as = "text")
id_1_httr_json <- fromJSON(id_1_httr)
id_1_httr_json
```
